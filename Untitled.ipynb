## TBML or trade-based money laundering is a methodology used to funnel illicit cash flows or illicit goods in and out of a country. It is the most misunderstood methodology of money laundering and there are no official estimates of the revenue it produces. The illegal economy that results from illicit trade and other illegal activity robs a country or given community from needed revenue and natural resources.  It allows illegal proceeds to be mixed with legal ones, providing a revenue stream to criminal organizations and terror groups.

## Imports

## Imports and exports by water are the most used transportation method. 800,000 tons were imported in 2020 by water and had the most value generated.  This includes ocean, river, Great Lakes, and intr-port shipments.  Next is air transport with less than 100,000 tons shipped and a value of almost 800,000 dollars was made. It looks like high value items were usually transported by water, air, and truck.  It also seems that less goods and capital flow was moved by foreign export than foreign import. 

```python
df.groupby('fr_inmode')['tons_2020'].sum().plot(kind='bar')
plt.xlabel('Shipping mode')
plt.ylabel('Tons')
plt.title('Tons shipped per mode of Import')
plt.show()
```

```python
df.groupby('fr_inmode')['value_2020'].sum().plot(kind='bar')
plt.xlabel('Shipping mode')
plt.ylabel('Value in Millions')
plt.title('Value per mode of Import')
plt.show()
```

## Exports

```python
df.groupby('fr_outmode')['tons_2020'].sum().plot(kind='bar')
plt.xlabel('Shipping mode')
plt.ylabel('Tons')
plt.title('Tons shipped per mode of Export')
plt.show()
```

```python
df.groupby('fr_outmode')['value_2020'].sum().plot(kind='bar')
plt.xlabel('Shipping mode')
plt.ylabel('Value in Millions')
plt.title('Tons shipped per mode of Export')
plt.show()
```

## Imports & Exports by country and value

## Eastern Asia imported the most goods and services via water followed by air. Even though air and water were the main modes of export for Eastern Asia, there is almost a 200,000 dollar difference between the two with more being spent on imports than made on exports.

```python
# Imports
plt.figure(figsize=(12,5))
df.groupby(['fr_orig','fr_inmode'])['value_2020'].sum().plot(kind='bar')
plt.xlabel('Region & Shipping mode')
plt.ylabel('Value')
plt.title('Expenses per mode of Import & Region')
plt.show()
```

```python
# Exports
plt.figure(figsize=(12,5))
df.groupby(['fr_dest','fr_outmode'])['value_2020'].sum().plot(kind='bar')
plt.xlabel('Region & Shipping mode')
plt.ylabel('Value')
plt.title('Capital gain per mode of Export & Region')
plt.show()
```

## There is a slight linear pattern where the value of freight increases as the weight increases. You see most outliers coming from Eastern Asia where there is a higher value but low weight. South and Central America has the greatest outlier of a value of 40,000 and less than 5000 tons.

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='tons_2020', y='value_2020', hue='fr_orig', data=df)
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
```
